````python
# TODO: متد زیر را برای رسم توزیع داده ها کامل کنید.
import matplotlib.pyplot as plt
import seaborn as sns

# TODO: متد برای رسم توزیع داده‌های عددی
def plot_numeric_distributions(df):
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

    # تنظیم اندازه نمودارها
    plt.figure(figsize=(15, 12))

    # رسم هیستوگرام برای هر ویژگی عددی
    for i, col in enumerate(numeric_cols):
        plt.subplot(len(numeric_cols) // 3 + 1, 3, i + 1)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f'توزیع داده‌های {col}')
        plt.xlabel(col)
        plt.ylabel('تعداد')

    plt.tight_layout()
    plt.show()

# TODO: بررسی هیستوگرام
plot_numeric_distributions(data)
````

````python
# TODO: داده های عددی و پرت را با باکس پلات بررسی کنید
import matplotlib.pyplot as plt
import seaborn as sns

def plot_numeric_boxplots(df):
    numeric_cols = df.select_dtypes(include=['number']).columns  # انتخاب ستون‌های عددی
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df[numeric_cols])
    plt.xticks(rotation=45)  # چرخش نام ستون‌ها برای خوانایی بهتر
    plt.title('Boxplots of Numeric Variables')
    plt.show()

# اجرای تابع با دیتافریم df
plot_numeric_boxplots(data)
````

````python
# TODO: شناسایی داده های پرت به کمک IQR

import numpy as np

def detect_outliers_iqr(df, columns):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers

numeric_columns = df_copy.select_dtypes(include=['number']).columns
outliers_iqr = detect_outliers_iqr(df_copy, numeric_columns)

````

````python
# TODO: شناسایی داده های پرت به کمک Z-Score

from scipy.stats import zscore

def detect_outliers_zscore(df, columns, threshold=3):
    outliers = {}
    for col in columns:
        z_scores = np.abs(zscore(df[col]))
        outliers[col] = df[z_scores > threshold]
    return outliers

outliers_zscore = detect_outliers_zscore(df_copy, numeric_columns)

````

````python
# TODO: ایمپورت کتابخانه های مورد نیاز

import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
````

````python
# TODO: متد زیر را کامل کنید.
def impute_missing_data(df, strategy_dict=None, auto_impute=True):
    """
    داده‌های گمشده رو با استراتژی‌های مشخص یا به صورت خودکار پر می‌کنه.

    پارامترها:
        df: دیتافریم ورودی
        strategy_dict: دیکشنری از ستون‌ها و استراتژی‌ها (مثلاً {'age': 'mean', 'job': 'mode'})
        auto_impute: اگر True باشد، داده‌ها به‌صورت خودکار پر می‌شوند.

    خروجی:
        دیتافریم پردازش شده
    """

    df_copy = df.copy()  # ایجاد کپی از دیتافریم اصلی

    if auto_impute:
        # پر کردن خودکار داده‌های عددی با میانگین و داده‌های دسته‌ای با مد
        for col in df_copy.columns:
            if df_copy[col].dtype == 'number':  # ستون عددی
                df_copy[col].fillna(df_copy[col].mean(), inplace=True)
            else:  # ستون دسته‌ای
                df_copy[col].fillna(df_copy[col].mode()[0], inplace=True)

    elif strategy_dict:
        # پر کردن بر اساس استراتژی مشخص شده
        for col, strategy in strategy_dict.items():
            if col in df_copy.columns:
                if strategy == 'mean':  # میانگین
                    df_copy[col].fillna(df_copy[col].mean(), inplace=True)
                elif strategy == 'median':  # میانه
                    df_copy[col].fillna(df_copy[col].median(), inplace=True)
                elif strategy == 'mode':  # مد
                    df_copy[col].fillna(df_copy[col].mode()[0], inplace=True)
                elif strategy == 'knn':  # الگوریتم KNN
                    imputer = KNNImputer(n_neighbors=5)
                    df_copy[col] = imputer.fit_transform(df_copy[[col]])
                elif strategy == "nearest":
                    df_copy[col].fillna(method="ffill", inplace=True)  # استفاده از مقدار قبلی
                elif isinstance(strategy, (int, float, str)):  # مقدار ثابت
                    df_copy[col].fillna(strategy, inplace=True)

    return df_copy  # بازگرداندن دیتافریم پردازش‌شده


````

````python
# TODO: اعمال متد روی دیتاست

custom_strategies = {
    "Transaction_Amount": "mean",
    "Transaction_Type": "mode",
    "Timestamp": "nearest",
    "Account_Balance": "median",
    "Device_Type": "mode",
    "Location": "mode",
    "Merchant_Category": "mode",
    "IP_Address_Flag": 0,
    "Previous_Fraudulent_Activity": 0,
    "Daily_Transaction_Count": "median",
    "Avg_Transaction_Amount_7d": "mean",
    "Failed_Transaction_Count_7d": "median",
    "Card_Type": "mode",
    "Card_Age": "mean",
    "Transaction_Distance": "median",
    "Authentication_Method": "mode",
    "Risk_Score": "mean",
    "Is_Weekend": "mode",
    "DayOfWeek": "mode",
    "HourOfDay": "mean",
    "WeekOfYear": "median",
    "Transaction_Ratio": "mean",
    "Failed_Transaction_Rate": "median",
    "Transaction_Amount_Ratio": "mean",
    "Fraud_Label": 0
}

df_processed = impute_missing_data(df_copy, strategy_dict=custom_strategies, auto_impute=False)

df_processed
````

````python
# TODO: متوازن سازی کلاس هدف

import pandas as pd
from collections import Counter
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTETomek, SMOTEENN

# بررسی توزیع کلاس‌ها
print("توزیع داده‌ها قبل از متوازن‌سازی:", Counter(y_train))

# انتخاب و اعمال بهترین روش
sampler = SMOTEENN()  # انتخاب SMOTEENN برای کاهش نویز و متوازن‌سازی بهتر
X_train_resampled, y_train_resampled = sampler.fit_resample(X_train, y_train)

print("توزیع داده‌ها پس از متوازن‌سازی:", Counter(y_train_resampled))
````

        همچنین عملکرد مدل ها را بررسی کنید و بهترین را برای مرحله بعدی یعنی بهینه سازی هایپرپارامتر ها انتخاب کنید.

````python
# TODO :  تعریف مدل‌هاو آموزش مدل ها
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score

# تعریف مدل‌ها
models = {
    "XGBoost": xgb.XGBClassifier(eval_metric="logloss"),
    "AdaBoost": AdaBoostClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "LightGBM": lgb.LGBMClassifier()
}

# آموزش و ارزیابی مدل‌ها
results = {}
for name, model in models.items():
    model.fit(X_train_resampled, y_train_resampled)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    results[name] = {"Accuracy": accuracy, "F1 Score": f1}
    print(f"{name}: Accuracy = {accuracy:.4f}, F1 Score = {f1:.4f}")



````

<br><br>
<div dir="rtl">
    <font size="4">
        <li> از مدل های آموزش داده شده بهترین مدل را انتخاب کنید تا ارزیابی های بیشتری انجام دهیم. </li>
    </font>
</div>
<br>

````python
# TODO: انتخاب بهترین مدل
# انتخاب بهترین مدل
best_model_name = max(results, key=lambda k: results[k]["F1 Score"])
best_model = models[best_model_name]
print(f"🔹 بهترین مدل انتخاب‌شده: {best_model_name}")
````

````python
# TODO: بررسی عملکرد مدل با Cross Validation

from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(best_model, X_train_resampled, y_train_resampled, cv=5, scoring="f1")

print(f"میانگین F1 Score در Cross Validation: {cv_scores.mean():.4f}")
````

````python
# TODO :  بررسی  عملکرد مدل روی داده های آموزش و تستس

from sklearn.metrics import confusion_matrix, classification_report

# پیش‌بینی روی داده‌های تست
y_pred_train = best_model.predict(X_train_resampled)
y_pred_test = best_model.predict(X_test)

# محاسبه Confusion Matrix
conf_matrix_train = confusion_matrix(y_train_resampled, y_pred_train)
conf_matrix_test = confusion_matrix(y_test, y_pred_test)

# نمایش گزارش عملکرد مدل
print("🔹 گزارش مدل روی داده‌های آموزشی:")
print(classification_report(y_train_resampled, y_pred_train))

print("🔹 گزارش مدل روی داده‌های تست:")
print(classification_report(y_test, y_pred_test))
````
